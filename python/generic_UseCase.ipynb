```python
import genericDataGrabber as DG
import numpy as np
```

```python
myfluid = DG.ThermoFluid("Water", "T", "P", "S", 100)
```

```python
myfluid.data["P"].min()
```

```python
myfluid.show_plot()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

newdata = myfluid.data[np.log10(myfluid.data["S"]) >= 1.0]
newdata = newdata[np.log10(newdata["S"])<=4.0]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")  # we want 3D plots
ax.scatter(newdata[myfluid.xvar], newdata[myfluid.yvar], np.log10(newdata[myfluid.zvar]), c=newdata[myfluid.zvar], cmap=myfluid.colorMap, edgecolors="none")  # Plot the data
# Set the Labels
ax.set_xlabel(myfluid.xvar)
ax.set_ylabel(myfluid.yvar)
ax.set_zlabel(myfluid.zvar)
ax.set_title("{0} and {1} vs {2} of {3}".format(
*myfluid.vars, myfluid.fluid))
plt.show()
```
